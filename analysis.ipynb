# 🎬 Netflix Content Analysis

This notebook analyzes the Netflix dataset to understand content distribution, production trends, and popular genres.

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os

# Set aesthetic styling
sns.set_theme(style="darkgrid", palette="muted")
plt.rcParams['figure.dpi'] = 120
plt.rcParams['font.family'] = 'sans-serif'
```

## 1. Load and Clean Data

```python
df = pd.read_csv('netflix.csv')

# Remove null values in important columns
df = df.dropna(subset=['type', 'country', 'release_year'])

# Ensure release_year is numeric and valid
df['release_year'] = pd.to_numeric(df['release_year'], errors='coerce').fillna(0).astype(int)
df = df[df['release_year'] > 2000] # Focus on modern content

df.head()
```

## 2. Movies vs TV Shows

```python
plt.figure(figsize=(8, 8))
type_counts = df['type'].value_counts()
colors = ['#E50914', '#221F1F']
plt.pie(
    type_counts, 
    labels=type_counts.index, 
    autopct='%1.1f%%', 
    colors=colors,
    explode=(0.05, 0),
    textprops={'color':"white", 'weight':'bold'}
)
plt.title("Movies vs TV Shows", fontsize=14, weight='bold')
plt.show()
```

## 3. Top Countries

```python
top_countries = df['country'].value_counts().head(10)
plt.figure(figsize=(12, 6))
sns.barplot(x=top_countries.index, y=top_countries.values, palette="Reds_r")
plt.xticks(rotation=45)
plt.title("Top 10 Content Producing Countries", fontsize=15, weight='bold')
plt.show()
```

## 4. Yearly Growth

```python
year_data = df['release_year'].value_counts().sort_index()
plt.figure(figsize=(12, 6))
plt.plot(year_data.index, year_data.values, color='#E50914', linewidth=3, marker='o')
plt.title("Growth of Content on Netflix", fontsize=15, weight='bold')
plt.show()
```

## 5. Top Genres

```python
genres = df['listed_in'].str.split(', ', expand=True).stack()
top_genres = genres.value_counts().head(10)
plt.figure(figsize=(12, 7))
sns.barplot(x=top_genres.values, y=top_genres.index, palette="flare")
plt.title("Top 10 Most Popular Genres", fontsize=15, weight='bold')
plt.show()
```
